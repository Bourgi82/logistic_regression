```python
import gradio as gr
import joblib
import numpy as np
import pandas as pd

# Charger le modèle et l'encodeur
model = joblib.load('RadomForest_model.pkl')
encoder = joblib.load("encoder.pkl")

# Définir les colonnes catégorielles
categorical_columns = [
    "sex", "is_smoking"
]

# Fonction de prétraitement
def preprocess_input(data):
    # Convertir les données en DataFrame
    input_df = pd.DataFrame([data])
    
    # Encoder les colonnes catégorielles
    encoded_features = encoder.transform(input_df[categorical_columns])
     # Combiner les caractéristiques encodées avec les numériques
    numerical_features = input_df.drop(columns=categorical_columns).values
    final_features = np.concatenate([numerical_features, encoded_features], axis=1)

    return final_features

# Fonction de prédiction
# Fonction de prédiction
def predict_gradio(
    id, age, education, cigsPerDay, BPMeds,
       prevalentStroke, prevalentHyp, diabetes, totChol, sysBP,
       diaBP, BMI, heartRate, glucose, TenyearCHD
):
    # Organiser les données d'entrée
    input_data = {
        "id": id,
        "age": age,
        "education": education,
        "cigsPerDay": cigsPerDay,
        "BPMeds": BPMeds,
        "prevalentStroke": prevalentStroke,
        "prevalentHyp": prevalentHyp,
        "diabetes": diabetes,
        "totChol": totChol,
         "sysBP": sysBP,
        "diaBP": diaBP,
        "BMI": BMI,
        "heartRate": heartRate,
        "glucose": glucose,
        "TenyearCHD": TenyearCHD,
            }

    # Prétraiter les données
    preprocessed_data = preprocess_input(input_data)

    # Faire une prédiction
    prediction = model.predict(preprocessed_data)

    # Retourner "yes" ou "no"
    return "yes" if prediction[0] == 1 else "no"

# Interface Gradio
inputs = [
    gr.Number(label="age"),
   gr.Dropdown(["admin", "technician", "services", "management", "unemployed", "retired", "blue-collar"], label="Job"),
    gr.Dropdown(["married", "single", "divorced"], label="Marital Status"),
    gr.Dropdown(["primary", "secondary", "tertiary", "unknown"], label="Education"),
    gr.Dropdown(["yes", "no"], label="Default"),
    gr.Number(label="Balance"),
    gr.Dropdown(["yes", "no"], label="Housing"),
    gr.Dropdown(["yes", "no"], label="Loan"),
    gr.Dropdown(["cellular", "telephone", "unknown"], label="Contact"),
    gr.Number(label="Day"),
    gr.Dropdown(["jan", "feb", "mar", "apr", "may", "jun", "jul", "aug", "sep", "oct", "nov", "dec"], label="Month"),
    gr.Number(label="Duration"),
    gr.Number(label="Campaign"),
    gr.Number(label="Pdays"),
    gr.Number(label="Previous"),
    gr.Dropdown(["success", "failure", "unknown", "other"], label="Poutcome"),
]

outputs = gr.Textbox(label="Prediction")

gr.Interface(
    fn=predict_gradio,
    inputs=inputs,
    outputs=outputs,
    title="Etude Médicale",
    description="Entrez les détails du client pour prédire s'il acceptera l'offre (yes) ou non (no).",
).launch()
```
